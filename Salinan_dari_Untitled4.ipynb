```python
import csv
import os
from datetime import datetime

print("Program nilai sederhana (bahasa Indonesia, gaya santai)")

nama = input("Masukkan nama: ").strip()

nim_input = input("Masukkan NIM (cuma angka ya): ").strip()
if nim_input.isdigit():
    nim = nim_input

    def ambil_nilai(teks):
        s = input(f"{teks} (0-100): ").strip()
        try:
            n = float(s)
        except ValueError:
            print(f"Input untuk {teks} nggak valid: harus angka.")
            return None
        if 0 <= n <= 100:
            return n
        else:
            print(f"Nilai untuk {teks} harus di antara 0 dan 100.")
            return None

    tugas1 = ambil_nilai("Tugas 1")
    tugas2 = ambil_nilai("Tugas 2")
    uts = ambil_nilai("UTS")
    uas = ambil_nilai("UAS")

    if tugas1 is None or tugas2 is None or uts is None or uas is None:
        print("Ada yang salah di input nilai. Jalankan ulang sel dan coba lagi ya.")
    else:
        nilai_akhir = 0.15 * tugas1 + 0.15 * tugas2 + 0.35 * uts + 0.35 * uas

        if nilai_akhir >= 90:
            nilai_huruf = "A"
        elif nilai_akhir >= 80:
            nilai_huruf = "AB"
        elif nilai_akhir >= 70:
            nilai_huruf = "B"
        elif nilai_akhir >= 60:
            nilai_huruf = "BC"
        elif nilai_akhir >= 50:
            nilai_huruf = "C"
        elif nilai_akhir >= 40:
            nilai_huruf = "D"
        elif nilai_akhir >= 30:
            nilai_huruf = "E"
        else:
            nilai_huruf = "F"

        status = "Lulus" if nilai_akhir >= 60 else "Tidak Lulus"

        print("\n=== Hasil ===")
        print(f"Nama: {nama}")
        print(f"NIM: {nim}")
        print(f"Nilai akhir: {nilai_akhir:.2f}")
        print(f"Grade: {nilai_huruf} — {status}")

        csv_file = "data_awal.csv"
        buat_header = os.path.exists(csv_file) == False
        with open(csv_file, "a", newline="", encoding="utf-8") as f:
            tulis = csv.writer(f)
            if buat_header:
                tulis.writerow(["timestamp","Nama","NIM","Tugas1","Tugas2","UTS","UAS","FinalScore","Grade","Status"])
            tulis.writerow([datetime.now().isoformat(), nama, nim, tugas1, tugas2, uts, uas, f"{nilai_akhir:.2f}", nilai_huruf, status])

        print(f"Data tersimpan ke {csv_file}")
else:
    print("NIM nggak valid: harus angka. Jalankan ulang sel ya.")
```

```python
import csv

# Nama file input dan output
file_input = 'data_awal.csv'
file_output = 'rekap_nilai_mhs.csv'

try:
    with open(file_input, mode='r', encoding='utf-8') as csvfile:
        reader = csv.DictReader(csvfile)

        # Mengambil header dari file input dan menambah kolom 'nilai_akhir'
        fieldnames = reader.fieldnames + ['nilai_akhir'] + [Tanggal ]

        with open(file_output, mode='w', newline='', encoding='utf-8') as outfile:
            writer = csv.DictWriter(outfile, fieldnames=fieldnames)
            writer.writeheader()

            for row in reader:
                # Konversi string ke float untuk perhitungan
                t1 = float(row['Tugas1'])
                t2 = float(row['Tugas2'])
                uts = float(row['UTS'])
                uas = float(row['UAS'])

                # Hitung nilai akhir sesuai bobot
                # (T1*10%) + (T2*10%) + (UTS*40%) + (UAS*20%)
                # Catatan: Total bobot Anda adalah 80%, pastikan ini sudah sesuai keinginan
                nilai_akhir = (t1 * 0.1) + (t2 * 0.1) + (uts * 0.4) + (uas * 0.2)

                # Tambahkan hasil perhitungan ke baris data
                row['nilai_akhir'] = nilai_akhir, 2

                # Tulis baris baru ke file output
                writer.writerow(row)

    print(f"Proses selesai. Hasil rekap disimpan di: {file_output}")

except FileNotFoundError:
    print(f"Error: File '{file_input}' tidak ditemukan.")
except Exception as e:
    print(f"Terjadi kesalahan: {e}")
```
