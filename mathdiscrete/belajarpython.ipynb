```python
def cari_x_kuadrat_4_ganjil():
    kandidat = []
    akar = [-2, 2]
    for x in akar:
        if x % 2 != 0: 
            kandidat.append(x)
    
    return set(kandidat) 
A = cari_x_kuadrat_4_ganjil()
print(f"A = {{x | x^2 = 4, x ganjil}}")
print(f"A = {A}")
print(f"Apakah A himpunan kosong? {len(A) == 0}")
print(f"A = ∅")
```

```python
def cari_x_kuadrat_lebih_dari_4():
    kandidat = []
    for x in range(-3, 4):
        if x**2 > 4:
            kandidat.append(x)
    return set(kandidat)
B = cari_x_kuadrat_lebih_dari_4()
print(f"B = {{x | x^2 > 4, x ∈ Z, -3 ≤ x ≤ 3}}")
print(f"B = {B}")
print(f"Apakah B himpunan kosong? {len(B) == 0}")
print(f"Jumlah elemen B: {len(B)}")
```

```python
def definisi_himpunan_universal():
    U = set(range(-5, 6))
    A = {x for x in U if x > 0}
    B = {x for x in U if x % 2 == 0}

    
    print(f"Himpunan Universal U = {U}")
    print(f"Himpunan A (bilangan positif) = {A}")
    print(f"Himpunan B (bilangan genap) = {B}")

    print(f"\nKomplemen A = {U - A}")
    print(f"Komplemen B = {U - B}")
    print(f"\nA ∩ B (irisan) = {A.intersection(B)}")
    print(f"A ∪ B (gabungan) = {A.union(B)}")
    
    print(f"\nApakah A ⊆ U? {A.issubset(U)}")
    print(f"Apakah B ⊆ U? {B.issubset(U)}")
definisi_himpunan_universal()
```

```python
def demonstrasi_himpunan_bagian():
    A = {1, 2, 3, 4, 5}
    B = {1, 2, 3}
    C = {2, 4, 6}
    D = {1, 2}
    
    print("Himpunan yang ada:")
    print(f"A = {A}")
    print(f"B = {B}")
    print(f"C = {C}")
    print(f"D = {D}\n")
    print("Hubungan subset (⊆):")
    print(f"Apakah B ⊆ A? {B.issubset(A)}")  
    print(f"Apakah C ⊆ A? {C.issubset(A)}")  
    print(f"Apakah D ⊆ B? {D.issubset(B)}") 
    print("\nHubungan proper subset (⊂):")
    print(f"Apakah D ⊂ B? {D.issubset(B) and D != B}")  
    print(f"Apakah B ⊂ B? {B.issubset(B) and B != B}") 
    
    print("\nHubungan superset (⊇):")
    print(f"Apakah A ⊇ B? {A.issuperset(B)}") 
    print(f"Apakah B ⊇ D? {B.issuperset(D)}") 
demonstrasi_himpunan_bagian()
```

```python
def demonstrasi_fungsi():

    X = {1, 2, 3, 4}
    Y = {1, 4, 9, 16} 
    
    print("1. Fungsi Kuadrat f(x) = x²")
    print(f"Domain (X) = {X}")
    print(f"Kodomain (Y) = {Y}")
    
    def f(x): return x**2
    
    print("\nPemetaan:")
    for x in X:
        print(f"f({x}) = {f(x)}")
    
    range_f = {f(x) for x in X}
    print(f"\nRange = {range_f}")
    
    def is_function_one_to_one():
        return len({f(x) for x in X}) == len(X)
    
    def is_function_onto():
        return range_f == Y
    
    print("\n2. Sifat Fungsi:")
    print(f"Apakah fungsi ini one-to-one (injektif)? {is_function_one_to_one()}")
    print(f"Apakah fungsi ini onto (surjektif)? {is_function_onto()}")
    print(f"Apakah fungsi ini bijektif? {is_function_one_to_one() and is_function_onto()}")
    
    print("\n3. Komposisi Fungsi g∘f dimana g(x) = √x")
    def g(x): return x**0.5
    
    print("(g∘f)(x) = g(f(x)):")
    for x in X:
        gf = g(f(x))
        print(f"(g∘f)({x}) = g({f(x)}) = {gf}")

demonstrasi_fungsi()
```

```python
def demonstrasi_fungsi_injektif():
    X = {0, 1, 2, 3} 
    def f(x): 
        return 2*x + 1
    print("Kasus: Fungsi f(x) = 2x + 1")
    print(f"Domain = {X}")
    print("\nPemetaan:")
    hasil_pemetaan = {}
    for x in X:
        hasil = f(x)
        hasil_pemetaan[x] = hasil
        print(f"f({x}) = {hasil}")
    def is_injective():
        hasil_fungsi = [f(x) for x in X]
        return len(set(hasil_fungsi)) == len(X)
    print("\nAnalisis Sifat Injektif:")
    print("1. Setiap elemen codomain harus berasal dari maksimal satu elemen domain")
    print("2. Tidak boleh ada dua input yang menghasilkan output yang sama")
    print(f"\nApakah fungsi ini injektif? {is_injective()}")
    if is_injective():
        print("\nPenjelasan: Fungsi ini injektif karena:")
        print("- f(0) = 1")
        print("- f(1) = 3")
        print("- f(2) = 5")
        print("- f(3) = 7")
        print("Setiap input menghasilkan output yang berbeda")
    else:
        print("\nPenjelasan: Fungsi ini tidak injektif karena ada output yang sama")
print("\nSoal Latihan:")
print("Diberikan fungsi g(x) = x² dengan domain {-2, -1, 0, 1, 2}")
print("Apakah fungsi tersebut injektif? Buktikan!")
demonstrasi_fungsi_injektif()
```

```python
def demonstrasi_fungsi_injektif_simple():
    def f(x): 
        return 2*x + 1
    domain = {1, 2, 3}
    print("Fungsi Injektif")
    print("Fungsi f(x) = 2x + 1")
    print(f"Domain = {domain}")
    
    print("\nHasil Pemetaan:")
    for x in domain:
        print(f"f({x}) = {f(x)}")
    hasil = [f(x) for x in domain]
    is_injektif = len(set(hasil)) == len(domain)
    print("\nPengecekan sifat injektif:")
    print(f"f(1) = {f(1)}")
    print(f"f(2) = {f(2)}")
    print(f"f(3) = {f(3)}")
    print(f"\nApakah fungsi ini injektif? {is_injektif}")
demonstrasi_fungsi_injektif_simple()
```

```python
def demonstrasi_fungsi_surjektif():
    X = {1, 2, 3, 4}  
    Y = {2, 3, 4, 5}  
    def f(x): 
        return x + 1
    print("Studi Kasus Fungsi Surjektif")
    print("Fungsi f(x) = x + 1")
    print(f"Domain (X) = {X}")
    print(f"Kodomain (Y) = {Y}")
    print("\nHasil Pemetaan:")
    range_f = set()
    for x in X:
        hasil = f(x)
        range_f.add(hasil)
        print(f"f({x}) = {hasil}")
    is_surjektif = range_f == Y
    print("\nPengecekan sifat surjektif:")
    print(f"Range f = {range_f}")
    print(f"Kodomain Y = {Y}")
    print(f"\nApakah fungsi ini surjektif? {is_surjektif}")
    if is_surjektif:
        print("\nPenjelasan: Fungsi ini surjektif karena:")
        print("Setiap elemen kodomain memiliki prapeta di domain:")
        for y in Y:
            prapeta = [x for x in X if f(x) == y]
            print(f"- {y} memiliki prapeta: {prapeta}")
    else:
        print("\nPenjelasan: Fungsi ini tidak surjektif")
        print("Ada elemen kodomain yang tidak memiliki prapeta")
demonstrasi_fungsi_surjektif()
```

```python
def demonstrasi_fungsi_surjektif_simple():
    X = {1, 2, 3} 
    Y = {2, 3, 4} 
    def f(x): return x + 1
    
    print("Fungsi Surjektif Sederhana")
    print(f"f(x) = x + 1")
    print(f"Domain = {X}")
    print(f"Kodomain = {Y}")
    range_f = {f(x) for x in X}
    print("\nPemetaan:")
    for x in X:
        print(f"f({x}) = {f(x)}")
    print(f"\nRange = {range_f}")
    print(f"Apakah surjektif? {range_f == Y}")
demonstrasi_fungsi_surjektif_simple()
```

```python
def demonstrasi_fungsi_bijektif():
    X = {1, 2, 3}
    Y = {2, 3, 4}
    
    def f(x): return x + 1
    print("Fungsi Bijektif")
    print("f(x) = x + 1")
    print(f"Domain = {X}")
    print(f"Kodomain = {Y}")
    range_f = {f(x) for x in X}
    is_injektif = len(range_f) == len(X)  
    is_surjektif = range_f == Y          
    is_bijektif = is_injektif and is_surjektif
    print("\nPemetaan:")
    for x in X:
        print(f"f({x}) = {f(x)}")
    print(f"\nRange = {range_f}")
    print("\nSifat fungsi:")
    print(f"Injektif? {is_injektif}")
    print(f"Surjektif? {is_surjektif}")
    print(f"Bijektif? {is_bijektif}")
    if is_bijektif:
        print("\nFungsi ini bijektif karena:")
        print("1. Bersifat injektif (one-to-one): setiap input menghasilkan output berbeda")
        print("2. Bersifat surjektif (onto): setiap elemen kodomain memiliki prapeta")
demonstrasi_fungsi_bijektif()
```

```python
def demonstrasi_fungsi_bijektif_simple():
    X = {1, 2, 3}
    Y = {2, 3, 4}
    def f(x): return x + 1
    print("Fungsi Bijektif Sederhana")
    print("f(x) = x + 1")
    print(f"Domain = {X}")
    print(f"Kodomain = {Y}")
    range_f = {f(x) for x in X}
    is_injektif = len(range_f) == len(X)
    is_surjektif = range_f == Y
    is_bijektif = is_injektif and is_surjektif
    print("\nPemetaan:")
    for x in X:
        print(f"f({x}) = {f(x)}")
    print(f"\nRange = {range_f}")
    print(f"Bijektif? {is_bijektif}")
    if is_bijektif:
        print("(Fungsi ini bijektif karena injektif dan surjektif)")
demonstrasi_fungsi_bijektif_simple()
```

```python
def demonstrasi_fungsi_beda_ukuran():
    X1 = {1, 2}       
    Y1 = {1, 2, 3, 4}  
    def f1(x): return x  
    print("Kasus 1: Domain < Kodomain")
    print(f"Domain X1 = {X1}")
    print(f"Kodomain Y1 = {Y1}")
    range_1 = {f1(x) for x in X1}
    print(f"Range = {range_1}")
    print(f"Injektif? {len(range_1) == len(X1)}")
    print(f"Surjektif? {range_1 == Y1}")
    print("→ Bisa injektif tapi tidak mungkin surjektif\n")
    
    # contoh 2: Domain lebih besar dr si Kodomainnya
    X2 = {1, 2, 3, 4}  
    Y2 = {1, 2}         
    def f2(x): return 1 if x <= 2 else 2
    print("Kasus 2: Domain > Kodomain")
    print(f"Domain X2 = {X2}")
    print(f"Kodomain Y2 = {Y2}")
    range_2 = {f2(x) for x in X2}
    print(f"Range = {range_2}")
    print(f"Injektif? {len(range_2) == len(X2)}")
    print(f"Surjektif? {range_2 == Y2}")
    print("→ Bisa surjektif tapi tidak mungkin injektif")
demonstrasi_fungsi_beda_ukuran()
```

```python
def demonstrasi_produk_kartesian_angka():
    A = {9, 4}
    B = {7, 6}
    produk_AB = {(a, b) for a in A for b in B}
    produk_BA = {(b, a) for b in B for a in A}
    print("Produk Kartesian dengan Bilangan")
    print(f"A = {A}")
    print(f"B = {B}")
    print(f"\nA × B = {produk_AB}")
    print(f"B × A = {produk_BA}")
    print(f"\nKardinalitas:")
    print(f"|A| = {len(A)}")
    print(f"|B| = {len(B)}")
    print(f"|A × B| = {len(produk_AB)} = {len(A)} × {len(B)}")
    print("\nA × B dalam bentuk daftar:")
    for pasangan in produk_AB:
        print(f"({pasangan[0]}, {pasangan[1]})")
demonstrasi_produk_kartesian_angka()
```

```python
def kartreltkomprelasi():
    A = {1, 2, 3}
    B = {2, 4, 6}
    C = {3, 6, 9}
    kartesian_AB = {(a,b) for a in A for b in B}
    print("1. Produk Kartesian")
    print(f"A = {A}")
    print(f"B = {B}")
    print(f"A × B = {kartesian_AB}")
    R1 = {(a,b) for a in A for b in B if b % a == 0}
    R2 = {(b,c) for b in B for c in C if c % b == 0}
    print("\n2. Relasi")
    print("R1 (a faktor dari b):")
    print(f"R1 = {R1}")
    print("\nR2 (b faktor dari c):")
    print(f"R2 = {R2}")
    komposisi = {(a,c) for a in A for c in C 
                if any(b for b in B if (a,b) in R1 and (b,c) in R2)}
    print("\n3. Komposisi Relasi")
    print("R2 ∘ R1 = {(a,c) | ∃b∈B, (a,b)∈R1 ∧ (b,c)∈R2}")
    print(f"R2 ∘ R1 = {komposisi}")
    print("\nPenjelasan komposisi:")
    for a, c in komposisi:
        path = [b for b in B if (a,b) in R1 and (b,c) in R2]
        print(f"{a} → {path} → {c}")
kartreltkomprelasi()
```
