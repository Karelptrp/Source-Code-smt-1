Map

```python
numbers = [1, 2, 3, 4, 5]
map_result = map(lambda x: x ** 2, numbers)
print(list(map_result))
```

```python
def kuadrat(x):
    return x * x

numbers = [1, 2, 3, 4, 5]

hasil_map = map(kuadrat, numbers)
print(list(hasil_map))


print(list(map(lambda x: x ** 2, numbers)))
```

```python
def kuadratlagi(x):
    return x * x

nimirs = [10, 20, 30, 40, 50]

hasil_map_kuadrat = map(lambda x: x ** 2, nimirs)
print(set(hasil_map_kuadrat))
hasil_map_lagi = list(map(kuadratlagi, nimirs))
print(hasil_map_lagi)
```
