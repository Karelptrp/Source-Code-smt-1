```python
#cara baru MODERN: WITH STATEMENT
with open("catatan1.txt", "w") as file:
    file.write("Halo Dunia")
```

Menulis ke file teks(.TXT)
- Pakai mode 'w'
- pakai method.write().
- penting: kita harus manual menambahkan enter (\n)

```python
lines=["Baris Satu\n", "Baris Dua\n", "Baris Tiga\n"]
with open("catatan2.txt", "w") as f:
    f.writelines(lines)
```

Membaca File Teks (.read())

- Membaca seluruh isi file sekaligus menjadi satu string panjang

```python
with open("demo.txt", "r") as f:
    isi = f.read()
    print(isi)
```

hanya bisa string, kalau set dan tuple gamau, kecuali pakai writelines

```python
line = ["Baris Satu\n", "Baris Dua\n", "Baris Tiga\n"]
with open("catatan5.txt", "w") as f:
    f.writelines(line)
```

```python
#perbedaan writelines dengan write adalah writelines dapat menulis banyak baris sekaligus sedangkan write hanya satu baris saja
#jika ingin menulis banyak baris dengan write, harus menggunakan perulangan
#contoh:
lines = ["Baris Satu\n", "Baris Dua\n", "Baris Tiga\n"]
with open("catatan6.txt", "w") as f:
    for line in lines:
        f.write(line)
#contoh di atas menggunakan write untuk menulis banyak baris dengan perulangan

writelines_list = ["Baris Empat\n", "Baris Lima\n", "Baris Enam\n"]
with open("catatan3.txt", "a") as f:
    f.writelines(writelines_list)

#jika write
write = ["Baris Empat\n", "Baris Lima\n", "Baris Enam\n"]
with open("catatan5.txt", "a") as f:
    for line in write:
        f.write(line)
#contoh di atas menggunakan write untuk menulis banyak baris dengan perulangan dan mode append
```

```python
writelines_list = ["Baris Empat\n", "Baris Lima\n", "Baris Enam\n"]
with open("catatan3.txt", "a") as f:
    f.writelines(writelines_list)
```
