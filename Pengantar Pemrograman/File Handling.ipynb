```python
#perbedaan writelines dengan write adalah writelines dapat menulis banyak baris sekaligus sedangkan write hanya satu baris saja
#jika ingin menulis banyak baris dengan write, harus menggunakan perulangan
#contoh:
lines = ["Baris Satu\n", "Baris Dua\n", "Baris Tiga\n"]
with open("catatan6.txt", "w") as f:
    for line in lines:
        f.write(line)
#contoh di atas menggunakan write untuk menulis banyak baris dengan perulangan

writelines_list = ["Baris Empat\n", "Baris Lima\n", "Baris Enam\n"]
with open("catatan3.txt", "a") as f:
    f.writelines(writelines_list)

#jika write
write = ["Baris Empat\n", "Baris Lima\n", "Baris Enam\n"]
with open("catatan5.txt", "a") as f:
    for line in write:
        f.write(line)
#contoh di atas menggunakan write untuk menulis banyak baris dengan perulangan dan mode append
```

W+ untuk write plus read and any operation

```python
line = ["Nama: Karel\n", "NIM: 23090250017\n"]
with open("biodata2.txt", "w+") as a:
    a.writelines(line)
    a.seek(0)#0 maka akan membaca dari awal atau semuanya, 1 maka akan membaca dari karakter ke 2 dst
    baca = a.read()
    print(baca)
```

membaca csv

```python
import csv

with open("data.csv", "w", newline='') as csvfile:
    fieldnames = ["Nama", "NIM", "Jurusan"]
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)

    writer.writeheader()
    writer.writerow({"Nama": "Kharel", "NIM": "23090250017", "Jurusan": "Informatika"})
    writer.writerow({"Nama": "Budi", "NIM": "23090250018", "Jurusan": "Sistem Informasi"})
```

```python
import csv

data = []
with open("data.csv", "r") as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        data.append(row)
```

PENGANTAR CSV (Comma Sheet Value)

Mengapa CSV penting untuk sains data
- Format universal pertukaran data
- Ringan (Hanya Teks)
- Dataset kaggle, data pemerintah, data ekspor sistem biasanya format .csv
- kita butuh modul bawaan: import csv.

```python
#Membaca CSV

import csv
with open("data.csv", "r") as f:
    baca = csv.reader(f)
    for baris in baca:
        print(baris)
```

```python
#file exception jika tidak ditemukan file csv
import csv
try:
    with open("data.csv", "r") as f:
        baca = csv.reader(f)
        for baris in baca:
            print(baris)
except FileNotFoundError:
    print("File 'data.csv' tidak ditemukan. Pastikan file sudah dibuat dengan menjalankan sel yang sesuai.")
```

```python
with open("hasil.csv", "r") as f:
    baca = csv.reader(f)
    header = next(baca, None)
    if header is None:
        print("File 'hasil.csv' kosong atau tidak memiliki header.")
    else:
        for baris in baca:
            nama = baris[0] if len(baris) > 0 else ""
            nilai = baris[1] if len(baris) > 1 else ""
            print(f"Nama: {nama}, Nilai: {nilai}")
```
