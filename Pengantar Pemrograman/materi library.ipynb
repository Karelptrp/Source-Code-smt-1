```python
KelilingLing = math.pi * 5 * 2
print(f"{KelilingLing: .2f}")

luasLingkaran = math.pi * math.pow(5,2)
print(f"{luasLingkaran:.2f}")

bulatA=math.ceil(luasLingkaran)
print(bulatA)

bulatB=math.floor(KelilingLing)
print(bulatB)
```
