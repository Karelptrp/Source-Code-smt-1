Try Except

```python
try:
    angka=int(input("Masukkan angka: "))
    print(10/ angka)
    pass
except ZeroDivisionError:
    print("Jangan memasukkan angka 0")
```

```python
angka=int(input("Masukkan angka: "))
print(f"angka yang dimasukkan adalah {angka}")
try:
    hasil=10/ angka
    print(f"Hasil pembagian 10 dengan {angka} adalah {hasil}")
    pass
except ZeroDivisionError:
    print("Jangan memasukkan angka 0")
```

```python
angka_a=int(input("Masukkan angka pertama: "))
angka_b=int(input("Masukkan angka kedua: "))
print(f"angka yang dimasukkan adalah {angka_a} dan {angka_b}")
try:
    hasil=angka_a / angka_b
    print(f"Hasil pembagian {angka_a} dengan {angka_b} adalah {hasil}")
    pass
except ZeroDivisionError:
     print("Jangan memasukkan angka 0 pada angka kedua")
except ValueError:
     print("Masukkan angka dengan format yang benar")
```

BLOK ELSE

blok ini kebalikan except, hanya berjalan jika try berhasil

```python
try:
    angka=int(input("Angka: "))
except ValueError:
    print("Input salah")
else:
    print(f"Anda memasukkan angka {angka}")
```

BLOK FINALLY
jika try berhasil dia muncul, jika try gagal dia tetap akan muncul

```python
try:
    angka=int(input("Masukkan angka: "))
    print(10/ angka)
except ValueError:
    print("Masukkan angka dengan format yang benar")
finally: #pasti dijalankan
    print("oalah")
```

Blok Raise(Membuat error sendiri)

sengaja membuat kondisi salah untuk pembanding
gunakan raise untuk melempar sebuah exception

```python
def hitung_gaji(hari):
    if hari <0:
        raise ValueError("Hari tidak boleh negatif")
    return hari * 150000

try:
    hari=int(input("Masukkan jumlah hari kerja: "))
    gaji=hitung_gaji(hari)
    print(f"Gaji Anda adalah {gaji}")
except ValueError as e:
    print(f"Error Logika {e}")
```
