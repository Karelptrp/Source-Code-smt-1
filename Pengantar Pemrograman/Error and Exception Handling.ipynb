Try Except

```python
try:
    angka=int(input("Masukkan angka: "))
    print(10/ angka)
    pass
except ZeroDivisionError:
    print("Jangan memasukkan angka 0")
```

```python
angka=int(input("Masukkan angka: "))
print(f"angka yang dimasukkan adalah {angka}")
try:
    hasil=10/ angka
    print(f"Hasil pembagian 10 dengan {angka} adalah {hasil}")
    pass
except ZeroDivisionError:
    print("Jangan memasukkan angka 0")
```

```python
angka_a=int(input("Masukkan angka pertama: "))
angka_b=int(input("Masukkan angka kedua: "))
print(f"angka yang dimasukkan adalah {angka_a} dan {angka_b}")
try:
    hasil=angka_a / angka_b
    print(f"Hasil pembagian {angka_a} dengan {angka_b} adalah {hasil}")
    pass
except ZeroDivisionError:
    if angka_b == 0:
     print("Jangan memasukkan angka 0 pada angka kedua")
except ValueError:
     print("Masukkan angka dengan format yang benar")
```

program diatas dapat dijalankan, namun jika menjalankan yang kedua akan error apabila dimasukkan string karena variabel angka_a dan angka_b harus didalem try.
