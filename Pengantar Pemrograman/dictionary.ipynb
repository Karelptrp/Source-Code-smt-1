```python
import datetime

MENU = {
    "MN001": ["Kopi Hitam", 12000],
    "MN002": ["Teh Tarik", 15000],
    "MN003": ["Donat", 10000],
    "MN004": ["Sandwich", 20000]
}


def tampilkan_menu():
    print("\nDaftar Menu Kafe")
    print("KODE | NAMA           | HARGA")
    print("-" * 30)
    for kode, (nama, harga) in MENU.items():
        print(f"{kode} | {nama:<13} | Rp {harga:,.0f}")
    print("-" * 30)


def kasir():
    pesanan = {}
    print("Selamat datang di Kasir Mini")

    while True:
        tampilkan_menu()
        inp = input("Masukkan kode dan jumlah (contoh: MN001,2) atau 'END': ")
        if inp.strip().upper() == 'END':
            break
        try:
            kode, jumlah = inp.split(',')
            kode = kode.strip().upper()
            jumlah = int(jumlah.strip())
            if kode not in MENU:
                print("Kode tidak ditemukan. Coba lagi.")
                continue
            if jumlah <= 0:
                print("Jumlah harus lebih dari 0.")
                continue
            pesanan[kode] = pesanan.get(kode, 0) + jumlah
            print(f"{jumlah}x {MENU[kode][0]} ditambahkan ke pesanan.")
        except ValueError:
            print("Format salah. Gunakan format KODE,JUMLAH (misal: MN001,2)")

    print("\n" + "#" * 40)
    print("STRUK PEMBELIAN")
    print(f"Tanggal: {datetime.date.today().strftime('%d-%m-%Y')}")
    print("#" * 40)
    if not pesanan:
        print("Tidak ada pesanan.")
        return

    total = 0
    print("QTY | KODE  | NAMA            | SUBTOTAL")
    print("-" * 40)
    for kode, qty in pesanan.items():
        nama, harga = MENU[kode]
        subtotal = harga * qty
        total += subtotal
        print(f"{qty:<3} | {kode} | {nama:<14} | Rp {subtotal:,.0f}")

    print("-" * 40)
    print(f"TOTAL: Rp {total:,.0f}")
    print("#" * 40)


if __name__ == '__main__':
    kasir()
```

```python
MENU = {
    "MN001": ("Kopi Hitam", 12000),
    "MN002": ("Teh Tarik", 15000),
    "MN003": ("Donat", 10000),
    "MN004": ("Sandwich", 20000)
}

pesanan_list = [] 
pesanan_set = set() 
TOKO = ("Kafe Mini", "Jl. Contoh No.1")


def tampilkan_menu():
    print("\nDaftar Menu (kode: nama - harga)")
    for kode, (nama, harga) in MENU.items():
        print(f"{kode}: {nama} - Rp {harga:,.0f}")


def tambahkan_pesanan(kode, jumlah=1):
    for _ in range(jumlah):
        pesanan_list.append(kode) 

    pesanan_set.add(kode) 


def hitung_total():
    total = 0
    for kode in set(pesanan_list):  
        qty = pesanan_list.count(kode) 
        nama, harga = MENU[kode]
        subtotal = qty * harga
        print(f"{qty}x {nama} - Rp {subtotal:,.0f}")
        total += subtotal
    return total


def contoh_tuple():
    nama_toko, alamat = TOKO
    print(f"Toko: {nama_toko} (Alamat: {alamat})")


def main():
    print("Program Kasir - Pembelajaran Struktur Data")
    contoh_tuple()
    tampilkan_menu()

    tambahkan_pesanan('MN001', 2)  
    tambahkan_pesanan('MN003', 1)  
    tambahkan_pesanan('MN002', 3) 

    print("\nIsi pesanan (list - urutan dan jumlah):", pesanan_list)
    print("Isi pesanan (set - kode unik):", pesanan_set)

    print("\nDetail pembelian:")
    total = hitung_total()
    print("-" * 30)
    print(f"TOTAL: Rp {total:,.0f}")


if __name__ == '__main__':
    main()
```
