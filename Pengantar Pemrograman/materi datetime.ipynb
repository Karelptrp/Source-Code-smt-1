```python
import datetime

saat_ini = datetime.datetime.now()
print(saat_ini)

print(saat_ini.year)
print(saat_ini.month)
print(saat_ini.day)
print(saat_ini.hour)
print(saat_ini.minute)
print(saat_ini.second)
print(saat_ini.microsecond)

```

```python
sekarang = datetime.datetime.now()
tampilan_bagus = sekarang.strftime("%A %d %B %Y %H:%M:%S")
print(tampilan_bagus)
```
